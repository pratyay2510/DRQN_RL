#### Plotting

```python
import pandas as pd
import matplotlib.pyplot as plt

drqn_file = 'cartpole_drqn_1000_s0.csv'
dqn_file = 'cartpole_dqn.csv'
dqn_10_file = 'cartpole_dqn_10.csv'

data_dqn = pd.read_csv(dqn_file)
data_drqn = pd.read_csv(drqn_file)
data_dqn_10 = pd.read_csv(dqn_10_file)

plt.figure(figsize=(10, 6))

plt.plot(data_dqn["Episode"], data_dqn["Mean_Reward"], label="DQN")
plt.plot(data_drqn["Episode"], data_drqn["Mean_Reward"], label="DRQN (Memory = 10)")
plt.plot(data_dqn_10["Episode"], data_dqn_10["Mean_Reward"], label="DQN (Memory = 10)")
plt.xlabel("Episodes")
plt.ylabel("Reward")
plt.title("Reward vs Episodes")
plt.legend()
plt.show()
```
